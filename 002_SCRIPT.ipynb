## STEP 2 - Demojize
#### 2.1. Sort emojis and assign unicode

```python
# Sort emojis
sorted_emojis = sorted(emoji_df["emoji"].unique(), key=lambda x: ord(x))

key = pd.DataFrame(sorted_emojis, columns=["emoji"])
key
```

```python
# Assign unicode
emoji_unicode_map = emoji_df.set_index('emoji')['unicode'].to_dict()
key['unicode'] = key['emoji'].map(emoji_unicode_map)

key
```

#### 2.2. Demojize

```python
# Demojize to assign quick overview in case the emoji is not legible by your browser

key["demojize"] = key["emoji"].apply(emoji.demojize)
key
```

#### 2.3. Replace
### OPTION 1: REPLACE ONLY THE TOP 10

```python
top10 = emoji_df["emoji"].value_counts().head(10)
top10
```

```python
# Initialize dictionary with empty values
#In this example, one of them (:melting_face:) has already been demojised

data10 = ["â™¥", "â˜º"]
keys10 = dict.fromkeys(data10, None)  

# Fill dictionary with unicode values
for e in data10:
    match = key.loc[key['emoji'] == e, 'demojize']
    if not match.empty:
        keys10[e] = match.iloc[0]  # assign the unicode value
    else:
        keys10[e] = None  # or some default if not found

print(keys10)
```

```python
# Map the dictionary to the dataframe
emoji_df["emoji"] = emoji_df["emoji"].map(keys10).combine_first(emoji_df["emoji"])
print(emoji_df["emoji"].value_counts().head(10))
```
